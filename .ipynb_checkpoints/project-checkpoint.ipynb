                                                      DATA LOADING

```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


df = pd.read_csv('amazon_delivery.csv')
df.head()
```

                                               DATA CLEANING

```python
df.isnull().sum()
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
import pandas as pd

# Select only numeric columns for IQR calculation
numeric_df = df.select_dtypes(include='number')

# Compute Q1, Q3, and IQR
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter rows: keep only those without outliers
df = df[~((numeric_df < lower_bound) | (numeric_df > upper_bound)).any(axis=1)]
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure only numeric columns are used
num_df = df.select_dtypes(include='number')

# Drop columns with no variance (corr is NaN)
num_df = num_df.loc[:, num_df.nunique() > 1]

# Compute correlation matrix safely
corr = num_df.corr()

# If everything is NaN (e.g., too few numeric cols), handle gracefully
if corr.isna().all().all():
    raise ValueError("Correlation matrix is all NaN. Check that you have at least two numeric columns with variance.")

# Optional: style and size
plt.figure(figsize=(10, 8))
sns.set_theme(style="white")

# Create heatmap
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5,
    square=True,
    cbar_kws={'shrink': 0.8}
)

plt.title("Correlation heatmap")
plt.tight_layout()
plt.show()
```

                                  FEATURE ENGINEERING

```python
X = df.drop('Delivery_Time', axis=1)
y = df['Delivery_Time']
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

# Keep only numeric columns
X_numeric = df.select_dtypes(include=['int64', 'float64'])

# Scale them
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X_scaled, y, test_size=0.2, random_state=42
)
```

                                      Model Training with Hyperparameter Tuning

1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression


param_lr = {
'C': [0.01, 0.1, 1, 10],
'solver': ['liblinear']
}


grid_lr = GridSearchCV(LogisticRegression(), param_lr, cv=5)
grid_lr.fit(X_train, y_train)
```

2) Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier


param_dt = {
'max_depth': [5, 10, 20],
'min_samples_split': [2, 5, 10]
}


grid_dt = GridSearchCV(DecisionTreeClassifier(), param_dt, cv=5)
grid_dt.fit(X_train, y_train)
```

3) Random Forest

```python
from sklearn.ensemble import RandomForestClassifier


param_rf = {
'n_estimators': [100, 200],
'max_depth': [10, 20],
'min_samples_split': [2, 5]
}


grid_rf = GridSearchCV(RandomForestClassifier(), param_rf, cv=5)
grid_rf.fit(X_train, y_train)
```

4) Support Vector Machine (SVM)

```python
from sklearn.svm import SVC


param_svm = {
'C': [0.1, 1, 10],
'kernel': ['linear', 'rbf']
}


grid_svm = GridSearchCV(SVC(), param_svm, cv=5)
grid_svm.fit(X_train, y_train)
```

5) Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB


nb = GaussianNB()
nb.fit(X_train, y_train)
```

6) KNearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier


param_knn = {
'n_neighbors': [3, 5, 7, 9],
'weights': ['uniform', 'distance']
}


grid_knn = GridSearchCV(KNeighborsClassifier(), param_knn, cv=5)
grid_knn.fit(X_train, y_train)
```

                                      Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

models = {
     'Logistic Regression': grid_lr,
     'Decision Tree': grid_dt,
     'Random Forest': grid_rf,
     'SVM': grid_svm,
     'Naive Bayes': nb,
     'KNN': grid_knn
}

for name, model in models.items():
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    print(name)
    print("MAE :", mae)
    print("MSE :", mse)
    print("RMSE:", rmse)
    print("R2 Score:", r2)
    print("-" * 30)
```

```python
from sklearn.metrics import r2_score
models = {
     'Decision Tree': grid_dt,
     'Random Forest': grid_rf,
     'SVM': grid_svm,
     'KNN': grid_knn
}


results = {}

for name, model in models.items():
    y_pred = model.predict(X_test)

    r2 = r2_score(y_test, y_pred)
    accuracy_percentage = r2 * 100
 
    results[name] = accuracy_percentage

    print(f"{name} Accuracy: {accuracy_percentage:.2f}%")

print("-" * 40)

# Best model
best_model = max(results, key=results.get)
print(f" Best Performing Model: {best_model}")
print(f" Accuracy Percentage: {results[best_model]:.2f}%")

```
